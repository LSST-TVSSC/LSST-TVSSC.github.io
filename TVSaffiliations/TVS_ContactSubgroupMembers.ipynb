```python
#just prints the emails of members of a group 
#both primary and secondary members

'__author__' == 'Federica Bianco, UDel - GitHub: fedhere'
import sys
import pandas as pd
import numpy as np
tvsfile = '1Q8YPGxh-Ebs02ykfGypj7oThQrgqKzrU6qyIuPT3bhY'
```

```python

sbuglist = ['']

TVSMembers = pd.read_csv('https://docs.google.com/spreadsheets/d/' +
                             tvsfile +
                             '/export?gid=0&format=csv')
TVSMembers.rename({"Unnamed: 0":"last name"}, axis=1, inplace=True)
#only select members that signed the CoC
TVSMembers = TVSMembers.loc[~TVSMembers["signed CoC"].isna()]
TVSMembers.head()
```

```python
#create the list of all subgroups in TVS
subgroups = (np.concatenate([TVSMembers.primary.unique(), 
                                     TVSMembers["secondary"].unique(), 
            TVSMembers["secondary.1"].unique(), 
            TVSMembers["secondary.2"].unique()]))

subgroups = list(sorted(set([s for s in subgroups if not isinstance(s, float)])))
subgroups
```

```python
import ipywidgets as widgets
from IPython.display import display
from IPython.html.widgets import interactive
```

```python

def alllist(x):
  print("subgroup selected:", x)
  # PRIMARY or SECONDARY
  emails = TVSMembers[(TVSMembers["primary"] == x) |
                      (TVSMembers["secondary"] == x) | 
                      (TVSMembers["secondary.1"] == x) | 
                      (TVSMembers["secondary.2"] == x)]['signed CoC'].values
  print(emails)
  return x

w = widgets.Select(options=subgroups)
interactive(alllist, x=w)
```

```python


def primarylist(x):
  print("primary selected:", x)
  # PRIMARY
  emails = TVSMembers[TVSMembers.primary == x]['email'].values
  print(emails)
  return x

w = widgets.Select(options=subgroups)
interactive(primarylist, x=w)
```

```python

def secondarylist(x):
  print("secondary selected:", x)
  # SECONDARY
  emails = TVSMembers[TVSMembers.secondary == x]['email'].values
  print(emails)
  return x

w = widgets.Select(options=subgroups)
interactive(secondarylist, x=w)
```
